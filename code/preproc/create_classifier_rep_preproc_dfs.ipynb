# Whole Brain

```python
activity_df_long = pd.read_csv(data_base_dir + 'schaefer_activity_long_df.csv')
```

### Activity

```python
wb_activity_df_long = activity_df_long.groupby(['sub_id', 'lith_word', 'var', 'study_block']).agg({'value': 'mean'}).reset_index()
wb_activity_df_long['roi'] = 'whole_brain'
wb_activity_df_long['var_group'] = 'whole_brain_schaefer'
```

### Pattern Similarity

```python
wb_pattern_corr_df = au.get_pattern_corr_df(activity_df_long, 
                    values='value', 
                    columns=['lith_word', 'study_block'],
                    group_index=['sub_id'], 
                    corr_index=['roi'],
                   lower_triangular=False)
wb_pattern_corr_df = wb_pattern_corr_df[wb_pattern_corr_df['study_block_1'] != wb_pattern_corr_df['study_block_2']]
wb_pattern_corr_df = wb_pattern_corr_df.drop(columns=['lith_word_study_block_1', 'lith_word_study_block_2'])    
```

```python
wb_pattern_corr_df['within_cross'] = 'Cross'
wb_pattern_corr_df.loc[wb_pattern_corr_df['lith_word_1'] == wb_pattern_corr_df['lith_word_2'], 'within_cross'] = 'Within'
wb_pattern_corr_df['fisher_z'] = np.arctanh(wb_pattern_corr_df['corr'])
```

```python
wb_pattern_corr_df
```

```python
wb_pattern_corr_df['min_study_block'] = wb_pattern_corr_df[['study_block_1', 'study_block_2']].min(axis=1)
wb_pattern_corr_df['max_study_block'] = wb_pattern_corr_df[['study_block_1', 'study_block_2']].max(axis=1)
wb_pattern_corr_df['study_block'] = wb_pattern_corr_df['study_block_1'].astype(str) + '_' + wb_pattern_corr_df['study_block_2'].astype(str)
wb_pattern_corr_df.loc[(wb_pattern_corr_df['within_cross'] == 'Within'), 'study_block'] = wb_pattern_corr_df['min_study_block'].astype(str) + '_' + wb_pattern_corr_df['max_study_block'].astype(str)
```

```python
wb_pattern_corr_df = wb_pattern_corr_df.groupby(['sub_id', 'lith_word_1', 'within_cross', 'study_block']).agg({'fisher_z': 'mean'}).reset_index()
```

```python
wb_pattern_corr_df = wb_pattern_corr_df.rename(columns={'lith_word_1': 'lith_word', 'fisher_z': 'value'})
wb_pattern_corr_df['var'] = 'pattern_sim_z_' + wb_pattern_corr_df['within_cross'].astype(str)
wb_pattern_corr_df = wb_pattern_corr_df.drop(columns=['within_cross'])
wb_pattern_corr_df['roi'] = 'whole_brain'
wb_pattern_corr_df['var_group'] = 'whole_brain_schaefer'
```

```python
wb_pattern_corr_df_long = wb_pattern_corr_df
```

### ISC

```python
pattern_df_pivot = pd.pivot_table(activity_df_long, 
                                      values='value', 
                                      index=['lith_word', 'study_block', 'roi'],
                                      columns=['sub_id']).reset_index()
wb_ISC_df_long = pattern_df_pivot.drop(columns=['roi']).groupby(
        ['lith_word', 'study_block']).corr().reset_index()
wb_ISC_melt = wb_ISC_df_long.melt(
        id_vars=['lith_word', 'study_block', 'sub_id'],
        value_vars=pattern_df_pivot.columns[3:], #subject cols
        var_name='sub_id_2',
        value_name='corr').dropna()
```

```python
wb_ISC_melt['fisher_z'] = np.arctanh(wb_ISC_melt['corr'])
wb_ISC_melt = wb_ISC_melt.query('sub_id != sub_id_2')
wb_ISC_df = wb_ISC_melt.groupby(['lith_word', 'sub_id', 'study_block']).agg({'fisher_z': 'mean'}).reset_index()
```

```python
wb_ISC_df = wb_ISC_df.rename(columns={'fisher_z': 'value'})
wb_ISC_df['var'] = 'ISC_z'
wb_ISC_df['roi'] = 'whole_brain'
wb_ISC_df['var_group'] = 'whole_brain_schaefer'
```

```python
wb_ISC_df_long = wb_ISC_df
```
